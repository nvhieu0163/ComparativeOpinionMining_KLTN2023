### Chuyển đổi file dữ liệu chỉ có nhãn CSI về định dạng chuẩn

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"total_23026_CSI_0.txt", sep = "\t", header=0)
```

```python
df["title"] = np.nan
df["subject"] = np.nan
df["object"] = np.nan
df["sức mạnh"] = np.nan
df["thiết kế"] = np.nan
df["giá"] = np.nan
df["tính năng"]= np.nan
df["an toàn"] = np.nan
```

```python
cols = list(df.columns.values)
print(cols)
```

```python
cols = ['sentence_idx', 'title', 'sentence', 'CSI', 'subject', 'object', 'sức mạnh', 'thiết kế', 'giá', 'tính năng', 'an toàn']
df_ = df[cols]
```

```python
df_['title'] =  df_['title'].fillna("none")
```

```python
df_['CSI'] = df_['CSI'].apply(np.int64)
```

```python
df_
```

```python
df_ = df_.sample(n=1000)
df_.to_csv(r"./txt/1000_CSI_0_noAC_noNER_.txt", header=True, index = None, sep="\t", mode='w')
```

```python
df_.to_excel("./xlsx/1000_CSI_0_noAC_noNER_.xlsx", index = None)
```
